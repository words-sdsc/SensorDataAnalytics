```python
import numpy as np
import pandas as pd 
from os import listdir
from os.path import isfile, join
from scipy.stats.stats import spearmanr
from scipy.stats.stats import pearsonr
%matplotlib inline
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sb
sb.set(style='ticks', palette='Set2')
```

```python
#read all seapearl files
perNodefiles = [f for f in listdir('.') if (isfile(join('.', f)) and 'pearl' in f)]
perNodefiles.sort()
perNodefiles
```

```python
sensors =["Tamb"] #, "CPU1", "MEM1", "CPU2", "MEM2", "ATX12", "ATX5", "ATX5sb", "ATX3.3", "HDD12", "HDD5", "Tina", "Tinb","Tcpu1", "Tcpu2", "Thca", "Tmem1", "Tmem2", "Tout"]
```

```python
#for each of the files, extract sensor data (=Tamb)
timeseriesdict = dict()
counter = 1
for nodeFile in [ 'pi14-seapearl.txt', 'pi08-seapearl.txt']:#perNodefiles:
    #read the file for this node
    print nodeFile
    df = pd.read_table(nodeFile, delim_whitespace=True)
    for givenSensor in sensors:       
            #create new dataframe for hashmap
            if (++counter) % 100 == 0:
                print counter
            #print "looking for :" + givenSensor
            view = df.loc[df['sensor'].isin([givenSensor])][0:160000]
            df_select = pd.DataFrame(view.values.copy(), view.index.copy(), view.columns.copy())
            #print df_select.shape
        
            #print number of data points in timeseries
            if (len(df_select['node']) == 0):
                continue
            KEY = (df_select['node'].iloc[0], df_select['sensor'].iloc[0]) #tuple
            #print nodeFile, givenSensor
            #print KEY
            
            del df_select['node']
            del df_select['sensor']
            
            #add to hashmap
            timeseriesdict[KEY] = df_select
            del df_select #only deletes the pointer to object
            #print '---------------------------'
     
    ##############################
    #the order of files printed gives label for each row from start of file
```

```python
import matplotlib.pyplot as plt
sb.set_context("notebook", rc={"lines.linewidth":4.5})
```

```python
# 1
fig =plt.figure(figsize=(14,12))
L = timeseriesdict[('pi14', 'Tamb')]
L['time']=pd.to_datetime(L['time'], infer_datetime_format=True)
plt.ylabel('Temperature at Node 14', fontsize=45)
plt.xlabel('~Time', fontsize=45)
plt.plot(L['value'][::1000])
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
fig.savefig('Final - Temp at Node 14.jpg')
```

```python
L = timeseriesdict[('pi14', 'Tamb')]
L['first_diff'] = L['value'] - L['value'].shift()
sum(L['first_diff'].iloc[1:])
```

```python
# 2
fig=plt.figure(figsize=(14,12))
L['first_diff'] = L['value'] - L['value'].shift()
plt.ylabel('Difference in Temperatue at Node 14', fontsize=45)
plt.xlabel('~Time', fontsize=45)
plt.plot(L['first_diff'][::1000])
plt.ylim(-1,1)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
fig.savefig('Final - Diff in Temperature -  Node 14.jpg')
```

```python
# 3
L = timeseriesdict[('pi14', 'Tamb')]
L['Lag in value'] = L['value'].shift()
ax=sb.jointplot('value', 'Lag in value', L[::100], kind='reg', size=20)
#ax.plot_joint(sns.kdeplot,cmap='Reds_d')
plt.ylabel('Temperature Shifted (past values)', fontsize=45)
plt.xlabel('Temperature (current values)', fontsize=45)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
ax.savefig('Final - Value vs Lag Node 14 every 100.png')
```

```python
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
```

```python
L = timeseriesdict[('pi14', 'Tamb')]
L['first_diff'] = L['value'] - L['value'].shift()
lag_correlations=acf(L['first_diff'].iloc[1::])
```

```python
lag_partial_correlations = pacf(L['first_diff'].iloc[1::])
```

```python
# 5
fig, ax = plt.subplots(figsize=(14,12))
sb.set_context("notebook", rc={"lines.linewidth":4.5})
plt.ylabel('Correlation Coefficient', fontsize=45)
plt.xlabel('Index of Past Temperature Value', fontsize=45)
ax.plot(lag_correlations, marker='o', linestyle='--', label='Auto Correlation')
ax.plot(lag_partial_correlations, marker='*', linestyle='--', label='Partial Auto Correlation')
ax.axes.legend(fontsize=30, loc='best')
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
fig.savefig('Final - FirstDiff_Correlation Coefficients Node 14.png')
```
