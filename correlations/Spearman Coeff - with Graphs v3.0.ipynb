```python
import numpy as np
import pandas as pd 
from os import listdir
from os.path import isfile, join
from scipy.stats.stats import spearmanr
from scipy.stats.stats import pearsonr
%matplotlib inline
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sb
```

```python
#read all seapearl files
perNodefiles = [f for f in listdir('.') if (isfile(join('.', f)) and 'pearl' in f)]
len(perNodefiles)
```

```python
sensors =["Tamb", "CPU1", "MEM1", "CPU2", "MEM2", "ATX12", "ATX5", "ATX5sb", "ATX3.3", "HDD12", "HDD5", "Tina", "Tinb","Tcpu1", "Tcpu2", "Thca", "Tmem1", "Tmem2", "Tout"]
```

```python
#for each of the files, extract sensor data (=Tamb)
timeseriesdict = dict()
counter = 1
for nodeFile in perNodefiles:
    #read the file for this node
    #print nodeFile
    df = pd.read_table(nodeFile, delim_whitespace=True)
    for givenSensor in sensors:       
            #create new dataframe for hashmap
            if (++counter) % 100 == 0:
                print counter
            #print "looking for :" + givenSensor
            view = df.loc[df['sensor'].isin([givenSensor])][0:160000]
            df_select = pd.DataFrame(view.values.copy(), view.index.copy(), view.columns.copy())
            #print df_select.shape
        
            #print number of data points in timeseries
            if (len(df_select['node']) == 0):
                #PRINTS <node,sensors> that are not recording
                print nodeFile, givenSensor
                continue
            
            KEY = (df_select['node'].iloc[0], df_select['sensor'].iloc[0]) #tuple
            #print nodeFile, givenSensor
            #print KEY
            
            del df_select['node']
            del df_select['sensor']
            
            #add to hashmap
            timeseriesdict[KEY] = df_select
            del df_select #only deletes the pointer to object
            #print '---------------------------'
     
    ##############################
    #the order of files printed gives label for each row from start of file
```

```python
L = timeseriesdict[('pi14', 'Tamb')]
L['value'].plot(figsize=(14,12))
```

```python
L['first_diff'] = L['value'] - L['value'].shift()
L['first_diff'].plot(figsize=(14,12))
sum(L['first_diff'].iloc[1:]) #mean value of diff series if close to zero
```

```python
L['sec_diff'] = L['first_diff'] - L['first_diff'].shift()
L['sec_diff'].plot(figsize=(14,12))
sum(L['sec_diff'].iloc[2:]) #mean value of diff series if close to zero
```

```python
L['Lag in diff'] = L['first_diff'].shift(1)
sb.jointplot('first_diff', 'Lag in diff', L, kind='reg', size=5)
```

```python
#auto correlated
L['Lag in value'] = L['value'].shift()
sb.jointplot('value', 'Lag in value', L, kind='reg', size=10)
```

```python
L['logged'] = L['value'].apply(lambda x: np.log(x))
L['logged'].plot(figsize=(14,12))
```

```python
L['logged first_diff'] = L['logged'] - L['logged'].shift()
L['logged first_diff'].plot(figsize=(14,12))
sum(L['logged first_diff'].iloc[1:]) #mean value of diff series if close to zero
```

```python
#auto correlated
L['Lag logged first_diff'] = L['logged first_diff'].shift()
sb.jointplot('logged first_diff', 'Lag logged first_diff', L, kind='reg', size=10)
L['Lag logged first_diff'].corr(L['logged first_diff'])
```

```python
otptfile = open('output.txt','w')
counter = 1
seen = set()
for key1, tsK in timeseriesdict.iteritems():
    for key2, tsQ in timeseriesdict.iteritems():
        if(key2 in seen):
            continue
        
        if (key1 == key2):
            continue
            
        if(counter%500 == 0):
            print counter, key1, key2, pcoeff
        
        tsK_timeformat = pd.to_datetime(tsK['time'], infer_datetime_format=True)
        tsK_Series = pd.Series(tsK['value'].tolist(), index = tsK_timeformat)
        j1=tsK_Series.resample('10S', closed='left')
        tsQ_timeformat = pd.to_datetime(tsQ['time'], infer_datetime_format=True)
        tsQ_Series = pd.Series(tsQ['value'].tolist(), index = tsQ_timeformat)
        j2=tsQ_Series.resample('10S', closed='left')
        J = j1.align(j2, join='inner')
        counter += 1
        
        #Difference
        L  = J[0] - J[0].shift()
        M  = J[1] - J[1].shift()
        
        #Compare
        if (len(J[0].tolist())==0):
            continue
        #pcoeff = pearsonr(J[0].values.tolist(), J[1].values.tolist())
        pcoeff = spearmanr(L.iloc[1:].values.tolist(), M.iloc[1:].values.tolist())
        if (pcoeff[0] >= 0.5 or pcoeff[0] <= -0.5):
            print '>>found...writing to file'
            otptfile.write(key1[0]+','+key1[1]+','+key2[0]+','+key2[1]+','+str(pcoeff[0])+','+str(pcoeff[1])+'\n')
            print key1[0]+','+key1[1]+','+key2[0]+','+key2[1]+','+str(pcoeff[0])+','+str(pcoeff[1])+'\n'
    seen.add(key1)
```
